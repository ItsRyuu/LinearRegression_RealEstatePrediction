```python
housing_fx = pd.read_csv('Rumah.csv')
housing_fx.head()
```

```python
varaible_list = ['price','guestroom','bedrooms','bathrooms','stories', 'parking']
housing_fx.head()
```

```python
df_train, df_test = train_test_split(housing_fx, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
df_train.shape
```

```python
scaler = MinMaxScaler()

#applying the scaler only to below variables
num_var= ['price', 'area', 'bedrooms', 'bathrooms', 'stories', 'parking']

df_train[num_var] = scaler.fit_transform(df_train[num_var])
```

```python
df_train.head()
```

```python
df_train.describe()
```

```python
y_train = df_train.pop('price')
X_train = df_train
```

```python
#Building Linear Model
import statsmodels.api as sm
```

```python
#area
X_train_sm = sm.add_constant(X_train[['area']])

lr_1 = sm.OLS(y_train, X_train_sm).fit()
```

```python
lr_1.params
```

```python
print(lr_1.summary())
```

```python
plt.scatter(X_train_sm.iloc[:,1], y_train)
plt.plot(X_train_sm.iloc[:,1],0.126894 + 0.462192*X_train_sm.iloc[:,1], 'r')
plt.show()
```

```python
#area and bathrooms

X_train_sm = sm.add_constant(X_train[['area', 'bathrooms']])

lr_2 = sm.OLS(y_train, X_train_sm).fit()
```

```python
lr_2.params
```

```python
print(lr_2.summary())
```

```python
X_train_sm = sm.add_constant(X_train[['area','bedrooms','bathrooms']])

lr_3 = sm.OLS(y_train, X_train_sm).fit()
```

```python
lr_3.params
```

```python
print(lr_3.summary())
```

```python
X_train_sm = sm.add_constant(X_train[['area','bedrooms','bathrooms','stories','parking']])
lr_4 = sm.OLS(y_train, X_train_sm).fit()
```

```python
lr_4.params
```

```python
print(lr_4.summary())
```

```python
X_train_sm = sm.add_constant(X_train[['area','bedrooms','bathrooms','stories','parking']])
lr_4 = sm.OLS(y_train, X_train_sm).fit()
```

```python
y_train_pred = lr_4.predict(X_train_sm)
```
